```python
# with open('d4_test.txt') as f:
with open('d4_input.txt') as f:
    d4input = f.readlines()

d4input
```

```python
draw = d4input[0]
draw = draw.split(',')
draw = [int(x) for x in draw]
print(draw)

boards = list()
for i in range(2, len(d4input), 6):
    boards.append(d4input[i:i+5])

def clean_boards(boards):
    for board in boards:
        for i in range(5):
            board[i] = board[i].split(' ')
            board[i] = [int(x) for x in board[i] if x]
    return boards

boards = clean_boards(boards)
boards
```

```python
def check_col(board, cur_draw):
    for i in range(5):
        if board[0][i] in cur_draw and \
            board[1][i] in cur_draw and \
            board[2][i] in cur_draw and \
            board[3][i] in cur_draw and \
            board[4][i] in cur_draw:
            return True
    return False


def check_row(board, cur_draw):
    for i in range(5):
        if board[i][0] in cur_draw and \
            board[i][1] in cur_draw and \
            board[i][2] in cur_draw and \
            board[i][3] in cur_draw and \
            board[i][4] in cur_draw:
            return True
    return False


def check_bingo(boards, cur_draw):
    winning_boards = list()
    for board in boards:
        if check_col(board, cur_draw) == True or \
            check_row(board, cur_draw) == True:
            winning_boards.append(board)
    return None if len(winning_boards) == 0 else winning_boards


winning_board = None
cur_draw = list()
i = -1

while winning_board is None and i < len(draw)-1:
    i += 1
    cur_draw.append(draw[i])
    bingos = check_bingo(boards, cur_draw)
    
    if bingos is None: winning_board = bingos
    else: winning_board = bingos[0]

cur_draw, winning_board
```

```python
def sum_board_points(board, cur_draw):
    board_sum = 0
    for i in range(5):
        for j in range(5):
            if board[i][j] not in cur_draw:
                board_sum += board[i][j]
    print(board)
    print(cur_draw)
    print(board_sum, cur_draw[len(cur_draw)-1])
    print(board_sum*cur_draw[len(cur_draw)-1])


sum_board_points(winning_board, cur_draw)
```

```python
last_winning_board = None
cur_draw = list()
i = -1
boards2 = boards.copy()

while len(boards2) >= 1 and i < len(draw)-1:
    i += 1
    cur_draw.append(draw[i])
    bingos = check_bingo(boards2, cur_draw)
    
    if bingos is None: last_winning_board = bingos
    else: last_winning_board = bingos[-1]
    
    if bingos is not None:
        for bingo in bingos:
            boards2.remove(bingo)
        

sum_board_points(last_winning_board, cur_draw)
```
